### Finding all the unique wind speed values in the data

### Renaming the column Weather to Weather Condition

```python
df.rename(columns = {"Weather": "Weather Condition"}, inplace = True)
df.head(2)
```

### Finding the mean Visibility

```python
df["Visibility_km"].mean()
```

### Standard Deviation of Pressure

```python
df["Press_kPa"].std()
```

 ### Finding all instances when wind speed is above 24 and visibility is 25

### Mean value of each column against each Weather Condition

```python
df.groupby("Weather Condition").mean()
```

### Minimum and maximum values of each column against each Weather Condition

```python
df.groupby("Weather Condition").max()
```

### Showing all the records where weather condition is fog
